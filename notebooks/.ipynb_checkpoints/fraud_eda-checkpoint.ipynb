```python
fraud_df['target'] = fraud_df.acct_type.str.contains('fraud')
fraud_df.target.value_counts()
```

```python
fraud_class = fraud_df.target.value_counts()/len(fraud_df.target)
fraud_class
```

```python
fig, ax = plt.subplots()

bars = ax.bar(x=['Not Fraud', 'Fraud'], height=fraud_class.values)
bars[0].set_color('#fbd334')
bars[1].set_color('#f84b52')
ax.set_facecolor('#f5f1e9')

for bar in bars:
    ax.annotate(f'{bar.get_height()*100:.0f}%', (bar.get_x()+ bar.get_width()/2 , bar.get_height()-0.01), 
                ha='center',fontsize=15)

ax.set_title('Fraud Class Proportion \n', fontsize=20)
fig.savefig('../images/fraud_class_prop.jpg');
```

```python
fraud_df.approx_payout_date = pd.to_datetime(fraud_df.approx_payout_date, unit='s')
fraud_df.event_created = pd.to_datetime(fraud_df.event_created, unit='s')
fraud_df.event_published = pd.to_datetime(fraud_df.event_published, unit='s')
fraud_df.event_start = pd.to_datetime(fraud_df.event_start, unit='s')
fraud_df.event_end = pd.to_datetime(fraud_df.event_end, unit='s')
fraud_df.user_created = pd.to_datetime(fraud_df.user_created, unit='s')
```

```python
time_cols = ['approx_payout_date', 'event_created', 'event_published', 'event_start', 'event_end' ]
fig, axs = plt.subplots(len(time_cols), figsize=(20,20))

for ax, col in zip(axs, time_cols):
    fraud_time =  fraud_df.loc[: , [col, 'target']]
    fraud_time = fraud_time.set_index(col)
    fraud_time = fraud_time.groupby(pd.Grouper(freq='M')).sum()
    ax.plot(fraud_time.index, fraud_time.target, color='#f84b52')
    ax.set_xlim(datetime.date(2010, 1, 1), datetime.date(2014, 1, 1))
```

```python
fraud_time =  fraud_df.loc[: , ['user_created', 'target']]
fraud_time = fraud_time.set_index('user_created')
fraud_time = fraud_time.groupby(pd.Grouper(freq='7d')).aggregate(['sum', 'count'])
fraud_time['proportion'] = fraud_time.target['sum']/fraud_time.target['count']
fraud_time

fig, axs = plt.subplots(2, figsize=(20,15))

axs[0].plot(fraud_time.index, fraud_time.target['count'], color='#1c60e8', linewidth=2)
axs[0].set_xlim(datetime.date(2009, 1, 1), datetime.date(2014, 1, 1))
axs[0].set_facecolor('#f5f1e9')
axs[0].set_title('\n Weekly Transaction Volume \n', fontsize=20)

axs[1].plot(fraud_time.index, fraud_time['proportion'], color='#f84b52', linewidth=3);
axs[1].set_xlim(datetime.date(2009, 1, 1), datetime.date(2014, 1, 1))
axs[1].set_facecolor('#f5f1e9')
axs[1].set_title('Proportion of Fraud Per Week \n', fontsize=20)
axs[1].axhline(0.1, color='black', ls='--', alpha=0.5)

fig.savefig('../images/fraud_timeline.jpg');
```

```python
fraud_time =  fraud_df.loc[: , ['approx_payout_date', 'target']]
fraud_time = fraud_time.set_index('approx_payout_date')
fraud_time = fraud_time.groupby(pd.Grouper(freq='7d')).aggregate(['sum', 'count'])
fraud_time['proportion'] = fraud_time.target['sum']/fraud_time.target['count']
fraud_time

fig, axs = plt.subplots(2, figsize=(20,15))

axs[0].plot(fraud_time.index, fraud_time.target['count'], color='#1c60e8', linewidth=2)
axs[0].set_xlim(datetime.date(2011, 1, 1), datetime.date(2014, 1, 1))
axs[0].set_facecolor('#f5f1e9')
axs[0].set_title('\n Weekly Transaction Volume \n', fontsize=20)

axs[1].plot(fraud_time.index, fraud_time['proportion'], color='#f84b52', linewidth=3);
axs[1].set_xlim(datetime.date(2011, 1, 1), datetime.date(2014, 1, 1))
axs[1].set_facecolor('#f5f1e9')
axs[1].set_title('Proportion of Fraud Per Week \n', fontsize=20)
axs[1].axhline(0.1, color='black', ls='--', alpha=0.5)
```

```python
fraud_by_country = fraud_df.loc[:,['country', 'target']].groupby('country').aggregate(['sum', 'count'])
fraud_by_country = fraud_by_country.droplevel(axis=1, level=[0])
fraud_by_country['diff'] = fraud_by_country['count'] - fraud_by_country['sum']
fraud_by_country = fraud_by_country.sort_values(by='count', ascending=False)
fraud_by_country.head()
```

```python
fig, ax = plt.subplots(figsize=(20,5))

ax.bar(fraud_by_country.index, fraud_by_country['diff']);
```

```python
fraud_by_country_top = fraud_by_country.nlargest(4, columns='diff')
fraud_by_country_top
```

```python
fraud_by_country_others = pd.DataFrame(fraud_by_country.drop(fraud_by_country_top.index).sum(), columns=['others']).T
fraud_by_country_others
```

```python
fraud_by_country_grouped = pd.concat([fraud_by_country_top, fraud_by_country_others])
fraud_by_country_grouped['fraud'] = fraud_by_country_grouped['sum'] / fraud_by_country_grouped['count']   
fraud_by_country_grouped['not_fraud'] = 1 - fraud_by_country_grouped['fraud'] 
fraud_by_country_grouped.head()
```

```python
fig, ax = plt.subplots(figsize=(20,5))

ax.bar(fraud_by_country_grouped.index, fraud_by_country_grouped['fraud'], label='Fraud', color='#f84b52')
ax.bar(fraud_by_country_grouped.index, fraud_by_country_grouped['not_fraud'], label='Not Fraud', color='#e6dcd7',
      bottom=fraud_by_country_grouped['fraud'])

country_labels = ['United States', 'Great Britain', 'Canada', 'Australia', 'Others']
ax.set_xticklabels(labels=country_labels)
ax.set_facecolor('#f5f1e9')
ax.legend(loc='upper left')
ax.set_title('\n Proportion of Fraud by Top Transaction Countries \n', fontsize=20)


for bar in ax.patches:
    ax.annotate(f'{bar.get_height()*100:.0f}%', (bar.get_x()+ bar.get_width()-0.4, bar.get_height()), 
                ha='center',fontsize=15, )
    

;
    
    
```

```python
fig, ax = plt.subplots(figsize=(8,8))

colors = ['#1c60e8', '#fbd334', '#64dbdb', '#9ab4c8' ,'#edabac']
explode=(0, 0, 0, 0, 0.1)
ax.pie(fraud_by_country_grouped['count'], labels=country_labels, autopct='%1.f%%', shadow=True,
      colors=colors, explode=explode, labeldistance=0.7)

ax.set_title('\n Transaction Volume by Country', fontsize=20);
```

```python
data_flattened = json_normalize(data,  'previous_payouts'
 , errors='ignore')
```

```python
fraud_prev_payout = fraud_df.previous_payouts.apply(lambda lst: len(lst))
fraud_prev_payout
```

```python
ids = []
for idx, payouts in zip(fraud_prev_payout.index, fraud_prev_payout.values):
    if payouts > 0:
        for _ in range(payouts):
            ids.append(idx)
```

```python
data_flattened['id'] = ids
data_flattened.head()
```

```python
fraud_flattened = fraud_df.copy()
fraud_flattened['id'] = fraud_flattened.index
fraud_flattened.head()
```

```python
fraud_merged = pd.merge(fraud_flattened, data_flattened, how='left', on='id')
fraud_merged.head(1)
```
